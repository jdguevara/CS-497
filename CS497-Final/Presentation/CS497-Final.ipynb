```python
import keras
import os
import numpy as np
from matplotlib import pyplot as plt

from keras.models import load_model
```

```python
from IPython.display import Image
```

# Load the Model

```python
loaded_model = load_model('FinalModel.h5')
```

# Prepare Images

```python
import numpy as np
import cv2
```

```python
def load_image(path):
    img = cv2.imread(path)
    img = cv2.resize(img,(150,150))
    img = np.reshape(img,[1,150,150,3])
    return img
```

# Make Predictions

* 0 - Not a Bronco B
* 1 - Bronco B

loaded_model.predict_classes(load_image('PathTo/Image'))

```python
BroncoBImagePath = "./Images/Yes/"
directory = os.fsencode(BroncoBImagePath)

for file in os.listdir(directory):
        filename = os.fsdecode(file)
        print(loaded_model.predict_classes(load_image(BroncoBImagePath + filename))[0][0])
        display(Image(BroncoBImagePath + filename))
```

```python
NotBroncoBImagePath = "./Images/No/"
directory = os.fsencode(NotBroncoBImagePath)

for file in os.listdir(directory):
        filename = os.fsdecode(file)
        print(loaded_model.predict_classes(load_image(NotBroncoBImagePath + filename))[0][0])
        display(Image(NotBroncoBImagePath + filename))
```
