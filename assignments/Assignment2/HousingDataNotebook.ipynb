```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
import sklearn
```

```python
data = pd.read_csv("housing_data.csv")
```

```python
data.head() # This will also print out the contents of data
```

```python
cols = ["GrLivArea", "LotArea", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd", "GarageCars", "1stFlrSF", "2ndFlrSF"]
data = data[cols]
```

```python
# Compute the "forward pass"
# For all training examples, compute predicted value given parameter values
preds = []
rows = data.shape[0]
cols = data.shape[1]
params = np.random.random(size = cols)

for i in range(rows):
    val = 0
    for j in range(cols):
        val = val + (data.iloc[i,j]) * params[j]
    preds.append(val)
preds

# This is not the most efficient way to do this, essentially we could cut down time by doing 
# matrix multiplication
```

```python
 # This gets us the same result as what we did with the previous loops but more efficiently
np.dot(data, params)
```
